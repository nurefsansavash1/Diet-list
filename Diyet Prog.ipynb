```python
import qrcode

    # Kullanıcıdan isim-soyisim, yaş, boy ve kilo bilgilerini alın.
class Kullanici:
    def __init__(self, isim, yas, boy, kilo):
        self.isim = isim
        self.yas = yas
        self.boy = boy
        self.kilo = kilo
        self.vki = kilo / (boy ** 2)
        self.kategori = self.hesapla_kategori()
        self.günlük_kalori = self.hesapla_günlük_kalori()

    def hesapla_kategori(self):
        if self.vki < 18.5:
            return "Zayıf"
        elif self.vki < 25:
            return "Normal kilolu"
        elif self.vki < 30:
            return "Fazla kilolu"
        elif self.vki < 40:
            return "Obez"
        else:
            return "İleri derecede obez"

    def hesapla_günlük_kalori(self):
        if self.vki < 18.5:
            return 3000
        elif self.vki < 25:
            return 2500
        elif self.vki < 30:
            return 2000
        elif self.vki < 40:
            return 1500
        else:
            return 1000

kullanicilar = []
for i in range(1):
    isim = input("İsim ve soyisim: ")
    yas = int(input("Yaş: "))
    boy = float(input("Boy (metre cinsinden): "))
    kilo = float(input("Kilo: "))
    kullanici = Kullanici(isim, yas, boy, kilo)
    kullanicilar.append(kullanici)


# Kullanıcının bilgilerini ve VKİ değerini ve sınıfını ekrana yazdırın.
#2f kullanımı, float veri tipindeki sayıların sadece 2 basamaklı ondalık kısmının görüntülenmesini sağlar.
print(f"{kullanici.isim} adlı kullanıcının VKİ değeri: {kullanici.vki:.2f}")
print(f"Sınıfı: {kullanici.kategori}")
print(f"Günlük alması gereken kalori miktarı: {kullanici.günlük_kalori}")

  # sınıfa göre kalori ihtiyacı verebiliriz. Kalori ihtiyacına göre diyet listesi verilcek . Kalori ihtiyaçlarını ortalama vericez ve önce cinsiyet ,kronik rahatsızlığı var mı sorusu sonra doktora yönlendirme 
# kronik rahatsızlığı yoksa diyet listesi verilcek . Diyet listesinde de ortalama yüzde 30 protein ,karbonhidrat gibi değerler seçtirilcek 
# Obez: 1500
#İleri derecede obez: 1000
#Fazla kilolu: 2000
#normal kilolu : 2500
#zayıf:3000

# diyet listesi vermeden önce 1 dk bekleyiniz çıktısı 
kronik_rahatsizlik= input("kronik rahatsızlığın var mı ? (Evet/Hayır): ")
#lower() metodu kullanarak kullanıcının verdiği cevabı küçük harflere dönüştürdük.
# Böylece, kullanıcının cevabı büyük harfle de olsa "evet" kelimesine karşılaştırılacaktır
if kronik_rahatsizlik.lower() == "evet":
    print("Size uygun bir diyet listesi vermemiz doğru olmaz QR kodda linkini bıraktığımız uzmana danışmanızı öneriyoruz.")

    # QR kodu oluşturma
    qr = qrcode.QRCode(version=1, box_size=10, border=5)
    qr.add_data('https://www.instagram.com/mutfaktakidiyetisyen/')
    qr.make(fit=True)

    # QR kodu gösterme
    img = qr.make_image(fill_color='black', back_color='white')
    img.show()

    # QR kodu dosyaya kaydetme
    img.save('example_qr_code.png')
elif kronik_rahatsizlik.lower()=="hayır":
    
    if vki < 18.5:
       print("Kilo Alma Diyeti:")
       print("- Kahvaltı: Yulaf ezmesi, süt ve muz")
       print("- Ara Öğün: Yoğurt ve fındık")
       print("- Öğle: Tavuk göğsü, kahverengi pirinç ve sebzeler")
       print("- Ara Öğün: Elma ve badem ezmesi")
       print("- Akşam: Izgara somon ve sebzeler")
    
    elif 18.5 <= vki<= 24.9:
      
      print("Sağlıklı Beslenme Diyeti:")
      print("- Kahvaltı: Yumurta, tam buğday ekmeği ve meyve")
      print("- Ara Öğün: Yoğurt ve meyve")
      print("- Öğle: Tavuk salatası, tam buğday ekmeği ve zeytinyağı")
      print("- Ara Öğün: Havuç ve humus")
      print("- Akşam: Izgara balık ve sebzeler")
    
    else:
      print("Kilo Verme Diyeti:")
      print("- Kahvaltı: Yumurta beyazı, tam buğday ekmeği ve meyve")
      print("- Ara Öğün: Sebzeler ve humus")
      print("- Öğle: Izgara tavuk göğsü, kahverengi pirinç ve sebzeler")
      print("- Ara Öğün: Badem ve kuru meyve")
      print("- Akşam: Izgara biftek ve sebzeler")
      
```
